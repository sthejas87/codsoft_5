```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

data = pd.read_csv('C:/Users/Sidharth/Downloads/creditcard.csv')
print("First few rows of the dataset:")
data.head()
```

```python
print("\nMissing values in the dataset:")
print(data.isnull().sum())
```

```python
X = data.drop(columns=['Class'])  # Features
y = data['Class']  # Target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nClass distribution in the original dataset:")
print(y.value_counts())
```

```python
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

model_orig = RandomForestClassifier(random_state=42)
model_orig.fit(X_train_orig, y_train_orig)

y_pred_orig = model_orig.predict(X_test_orig)
print("\nPerformance on Original Imbalanced Dataset:")
print(classification_report(y_test_orig, y_pred_orig))
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

print("\nClass distribution after SMOTE:")
print(y_resampled.value_counts())
```

```python
X_train_bal, X_test_bal, y_train_bal, y_test_bal = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

model_bal = RandomForestClassifier(random_state=42)
model_bal.fit(X_train_bal, y_train_bal)

y_pred_bal = model_bal.predict(X_test_bal)
print("\nPerformance on Balanced Dataset (after SMOTE):")
print(classification_report(y_test_bal, y_pred_bal))
```
